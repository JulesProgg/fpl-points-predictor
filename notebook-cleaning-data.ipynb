```python
USEFUL_COLS = [
    # Identification
    "id", "name", "team", "position",

    # Raw statistics
    "minutes", "goals_scored", "assists",
    "goals_conceded", "clean_sheets", "saves",
    "penalties_saved", "penalties_missed",
    "yellow_cards", "red_cards", "own_goals",
    "starts", "bonus", "bps",

    # Advanced cumulative statistics
    "expected_goals", "expected_assists",
    "expected_goal_involvements",  # Don't have for 24-25
    "expected_goals_conceded",
    "influence", "creativity", "threat", "ict_index",

    # Target
    "total_points",
]
```

```python
df22 = df22[[c for c in USEFUL_COLS if c in df22.columns]]
df23 = df23[[c for c in USEFUL_COLS if c in df23.columns]]
df24 = df24[[c for c in USEFUL_COLS if c in df24.columns]]
```

```python
import pandas as pd

base_path = "/files/fpl-points-predictor/data/raw"

df22 = pd.read_csv(base_path + "/season22-23.csv")
df23 = pd.read_csv(base_path + "/season23-24.csv")
df24 = pd.read_csv(base_path + "/season24-25.csv")
```

```python
# 1) Renommer les colonnes spéciales de 24-25
rename_24 = {
    "first_name": "first",
    "second_name": "second",
    "player_position": "position",
    "team_name": "team",
}

df24 = df24.rename(columns=rename_24)

# 2) Construire une colonne 'name' à partir de 'first' + 'second'
df24["name"] = df24["first"].str.strip() + " " + df24["second"].str.strip()

# 3) On peut ensuite supprimer 'first' et 'second' si on veut
df24 = df24.drop(columns=["first", "second"])
```

```python
print("df22:", [c for c in df22.columns if c in ["name", "team", "position"]])
print("df23:", [c for c in df23.columns if c in ["name", "team", "position"]])
print("df24:", [c for c in df24.columns if c in ["name", "team", "position"]])
```

```python
def select_useful(df):
    cols = [c for c in USEFUL_COLS if c in df.columns]
    return df[cols].copy()
```

```python
df22_useful = select_useful(df22)
df23_useful = select_useful(df23)
df24_useful = select_useful(df24)
```
