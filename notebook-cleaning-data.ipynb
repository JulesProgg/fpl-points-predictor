```python
"""
data_pipeline.py

Data cleaning and preparation pipeline for the FPL Points Predictor project.

Steps:
- Load the three raw CSV files (one per season) from data/raw/
- Harmonise column names across seasons
- Keep only the relevant columns (USEFUL_COLS)
- Add a 'season' column
- Concatenate all seasons into a single dataset
- Save the processed dataset to data/processed/players_all_seasons.csv
"""

from pathlib import Path
from typing import List, Tuple

import pandas as pd

# --- Paths -------------------------------------------------------------------

# Assuming this file is in src/, project root is one level above

PROJECT_ROOT = Path("/files/fpl-points-predictor")
DATA_RAW_DIR = PROJECT_ROOT / "data" / "raw"
DATA_PROCESSED_DIR = PROJECT_ROOT / "data" / "processed"

# Filenames and season labels
SEASONS_CONFIG: List[Tuple[str, str]] = [
    ("season22-23.csv", "2022-23"),
    ("season23-24.csv", "2023-24"),
    ("season24-25.csv", "2024-25"),
]

# --- Columns to keep ---------------------------------------------------------

USEFUL_COLS = [
    # identification
    "id", "name", "team", "position",

    # raw statistics
    "minutes", "goals_scored", "assists",
    "goals_conceded", "clean_sheets", "saves",
    "penalties_saved", "penalties_missed",
    "yellow_cards", "red_cards", "own_goals",
    "starts", "bonus", "bps",

    # advanced cumulative statistics
    "expected_goals", "expected_assists",
    "expected_goal_involvements",  # not available for 24-25
    "expected_goals_conceded",
    "influence", "creativity", "threat", "ict_index",

    # target
    "total_points",
]


# --- Helper functions --------------------------------------------------------

def normalise_columns(df: pd.DataFrame) -> pd.DataFrame:
    """
    Harmonise column names across seasons:
    - build 'name' from 'first_name' and 'second_name' if needed
    - rename player_position -> position
    - rename team_name -> team
    """
    df = df.copy()

    # Build "name" if needed and possible
    if "name" not in df.columns:
        if "first_name" in df.columns and "second_name" in df.columns:
            df["name"] = (
                df["first_name"].astype(str).str.strip()
                + " "
                + df["second_name"].astype(str).str.strip()
            )
            # We can drop the original name parts if they exist
            df = df.drop(columns=["first_name", "second_name"])

    # Rename columns to match the others
    rename_map = {}
    if "player_position" in df.columns and "position" not in df.columns:
        rename_map["player_position"] = "position"
    if "team_name" in df.columns and "team" not in df.columns:
        rename_map["team_name"] = "team"

    if rename_map:
        df = df.rename(columns=rename_map)

    return df


def select_useful(df: pd.DataFrame) -> pd.DataFrame:
    """
    Return a new DataFrame containing only the relevant columns
    defined in USEFUL_COLS (keeping only those that exist in df).
    """
    cols = [c for c in USEFUL_COLS if c in df.columns]
    return df[cols].copy()


def load_and_clean_season(filename: str, season_label: str) -> pd.DataFrame:
    """
    Load one season from data/raw, normalise columns, keep useful columns,
    and add a 'season' column.
    """
    path = DATA_RAW_DIR / filename
    if not path.exists():
        raise FileNotFoundError(f"Raw data file not found: {path}")

    df = pd.read_csv(path)
    df = normalise_columns(df)
    df = select_useful(df)

    # Add season label (useful later for train/test splits)
    df["season"] = season_label

    return df


def build_full_dataset() -> pd.DataFrame:
    """
    Load, clean and concatenate all seasons defined in SEASONS_CONFIG.
    """
    season_dfs = [
        load_and_clean_season(filename, season_label)
        for filename, season_label in SEASONS_CONFIG
    ]
    full = pd.concat(season_dfs, ignore_index=True)
    return full


def save_processed_dataset(df: pd.DataFrame,
                           filename: str = "players_all_seasons.csv") -> Path:
    """
    Save the processed dataset to data/processed/.
    """
    DATA_PROCESSED_DIR.mkdir(parents=True, exist_ok=True)
    out_path = DATA_PROCESSED_DIR / filename
    df.to_csv(out_path, index=False)
    return out_path


def run_pipeline() -> Path:
    """
    Run the full data pipeline:
    - build the full dataset
    - save it to data/processed/

    Returns:
        Path to the saved CSV file.
    """
    full_df = build_full_dataset()
    out_path = save_processed_dataset(full_df)
    return out_path


if __name__ == "__main__":
    output = run_pipeline()
    print(f"Data pipeline completed. File saved to: {output}")
```

```python
"""
data_pipeline.py

Simple data cleaning pipeline for the FPL Points Predictor project.

Steps:
- Load the three raw CSV files (one per season) from data/raw/
- Normalise the 2024-25 dataset so columns match the others
- Keep only the relevant columns (USEFUL_COLS)
- Concatenate all three seasons
- Save the final dataset to data/processed/players_all_seasons.csv
"""

from pathlib import Path
import pandas as pd

# Create variable to have quick path access
PROJECT_ROOT = Path("/files/fpl-points-predictor")
DATA_RAW_DIR = PROJECT_ROOT / "data" / "raw"
DATA_PROCESSED_DIR = PROJECT_ROOT / "data" / "processed"



# Relevant columns from the raw datasets
USEFUL_COLS = [
    # identification
    "id", "name", "team", "position",

    # raw statistics
    "minutes", "goals_scored", "assists",
    "goals_conceded", "clean_sheets", "saves",
    "penalties_saved", "penalties_missed",
    "yellow_cards", "red_cards", "own_goals",
    "starts", "bonus", "bps",

    # advanced cumulative statistics
    "expected_goals", "expected_assists",
    "expected_goal_involvements",  # not available for 24-25
    "expected_goals_conceded",
    "influence", "creativity", "threat", "ict_index",

    # target
    "total_points",
]

# create useful functions
def select_useful(df: pd.DataFrame) -> pd.DataFrame:
    """
    Return a new and independnant DataFrame containing only the relevant columns
    defined in USEFUL_COLS (keeping only those that exist in df).
    """
    cols = [c for c in USEFUL_COLS if c in df.columns]
    return df[cols].copy()


def run_pipeline() -> Path:
    """
    Run the full pipeline:
    - load the three raw datasets
    - normalise the 2024-25 dataset
    - select useful columns
    - concatenate all seasons
    - save the final dataset to data/processed/

    Returns:
        Path to the saved CSV file.
    """

    # Load the 3 raw season datasets
    df22 = pd.read_csv(DATA_RAW_DIR / "season22-23.csv")
    df23 = pd.read_csv(DATA_RAW_DIR / "season23-24.csv")
    df24 = pd.read_csv(DATA_RAW_DIR / "season24-25.csv")

    # Normalise the 2024-25 dataset so that it matches the others 

    # 1) Rename columns to match the other seasons
    rename_24 = {
        "first_name": "first",
        "second_name": "second",
        "player_position": "position",
        "team_name": "team",
    }
    df24 = df24.rename(columns=rename_24)

    # 2) Build a "name" column from "first" and "second"
    df24["name"] = df24["first"].str.strip() + " " + df24["second"].str.strip()

    # 3) Drop the temporary columns
    df24 = df24.drop(columns=["first", "second"])

    # Keep only useful columns for each season
    df22_useful = select_useful(df22)
    df23_useful = select_useful(df23)
    df24_useful = select_useful(df24)

    # Concatenate the three seasons 
    full = pd.concat([df22_useful, df23_useful, df24_useful], ignore_index=True)

    # Save the final dataset 
    DATA_PROCESSED_DIR.mkdir(parents=True, exist_ok=True)
    out_path = DATA_PROCESSED_DIR / "players_all_seasons.csv"
    full.to_csv(out_path, index=False)

    return out_path


if __name__ == "__main__":
    output_path = run_pipeline()
    print(f"Data pipeline completed. File saved to: {output_path}")
```
